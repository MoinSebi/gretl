```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
# File
f = "../data/test/yeast/analysis/yeast.chr5.core.pansn.sim.txt"
```

```python
# Read the table
df = pd.read_csv(f, sep = "\t", index_col = 0)
```

```python
# Read the private sequence
f2 =  "../data/test/yeast/analysis/yeast.chr5.core.pansn.sim.private.txt"

df_seq = pd.read_csv(f2, sep = "\t", index_col = 0)
```

```python
# Sort value
df_seq2 = df_seq.sort_values("Sequence[bp]")
df_seq2
```

```python
# Print the bar plot (sequence)
plt.bar([int(x) for x in df.index], df["Sequence[bp]"], edgecolor = "black")
plt.bar([int(x) for x in df.index][-1], df["Sequence[bp]"].values[-1], edgecolor = "black")

bottom1 = 0
for x in df_seq2.iterrows():
    plt.bar(1, (x[1]["Sequence[bp]"]), color = "yellow", bottom = bottom1, edgecolor = "black")
    bottom1 += int(x[1]["Sequence[bp]"])
plt.ylabel("Sequence [bp]")
plt.xlabel("Similarity (Number of samples)")
#plt.savefig("plots/pancore.pdf")
```
