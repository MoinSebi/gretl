```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = "../data/test/yeast/stats/stats.path2.tsv"
```

```python
df = pd.read_csv(f, sep = "\t")
```

```python
df.index = df["Path"]
```

```python
df
```

```python
df.index = [x.split(".")[0] for x in df.index]
```

```python
import matplotlib as mpl
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False
```

```python
plt.figure(figsize = (5,4))
plt.scatter(data=df, x="Edges", y="Jumps total", edgecolor = "black", color = "royalblue")
for x in df.iterrows(): 
    plt.annotate(x[0], (x[1]["Edges"] + 0.0005, x[1]["Jumps total"]))
plt.ylabel("Edges")
plt.xlabel("Jumps total")
plt.tight_layout()
#plt.savefig("path.scatter.png", dpi = 400)
```
