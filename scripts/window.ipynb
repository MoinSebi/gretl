```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Example file
f = "../data/test/yeast/analysis/analysis.window.txt"
```

```python
# Read the file using pandas
df = pd.read_csv(f, sep = "\t", header = None)
```

```python
# Set new index
df.set_index(0, inplace=True)
```

```python
# Show the dataframe that everything looks how it is supposed to
df
```

```python
# Plot the figure
plt.figure(figsize=(10, 6))  # Set the figure size
sns.heatmap(df, cmap="coolwarm", cbar_kws={'label': 'Similarity'})
plt.tight_layout(rect=[0.05, 0.05, 0.85, 0.95])  # Adjust as needed
plt.xlabel("Sequence window [x 5 kb]")
#plt.savefig("plots/analysis.window.pdf")
```
